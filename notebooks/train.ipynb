```python
import pexpect
import sys

# Spawn the process with a pseudo-TTY
child = pexpect.spawn('gh auth login --web', encoding='utf-8')

# Redirect child output to your terminal so you can see everything live
child.logfile = sys.stdout

# Wait for the expected prompt
child.expect('Authenticate GitHub CLI with your GitHub account.*')

# Send 'Y' to confirm
child.sendline('Y')

# Hand control back to the user (for browser interaction)
child.interact()
```
